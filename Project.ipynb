```python
%matplotlib inline
import pandas as pd
```

```python
data = pd.read_csv('../invoices.csv', parse_dates=[1],
                     index_col = 'date')
data
```

```python
data2 = pd.read_csv('../invoices2.csv', parse_dates=[1],
                     index_col = 'date')
data2 = data2.ix['2012-07-01':'2014-06-30']
```

```python
#merge the two csv files
data = data.append(data2)
```

```python
#Goal: make new dataframe of all orders - Caution this can take a while
import json
test = json.loads(data.contents[0])
#Add the first invoice
df_orders = pd.DataFrame(test)
df_orders['order_id'] = data.id[0]
df_orders['cust_id'] = data.cust_id[0]
#Add the rest of the invoices
for i in range (1, len(data)-1):
    orders = json.loads(data.contents[i])
    temp_df = pd.DataFrame(orders)
    #add order id & customer id
    temp_df['order_id'] = data.id[i]
    temp_df['cust_id'] = data.cust_id[i]
    df_orders = df_orders.append(temp_df)
```

```python
df_orders.reset_index()
```

```python
df_orders = df_orders[['cust_id', 'discount', 'id', 'name', 'order_id', 'price', 'qty', 'rowid', 'subtotal']]
```

```python
df_orders
```

```python
products = df_orders.groupby('name')
product_series = pd.Series(products.size())
all_products = pd.DataFrame(product_series)
len(all_products)
```

```python
#drop the first two rows, not relevant
all_products = all_products.drop(all_products.index[[0, 1]])
all_products=all_products.reset_index()
print len(all_products)
all_products.head()
```

```python
all_products.columns = ['product_name', 'no_sold']
all_products.head()
```

```python
import re
#the numbers in the product name represent colours. Can I extract this out to a new column?
productnames = all_products.product_name
colours = []
for i in range(0, len(productnames)):
    name = productnames[i]
    #get all product names that contain a colour
    if "-" in name:
        #extract the colour
        if re.search(r'-\d.*', name) is not None:
            match = re.search(r'-\d.*', name)
            c = match.group()
            colour = c[1:]
            code = -1
            if colour is not None: 
                digit = colour.encode('utf-8')
                code = int(digit[0])
            colours.append(code)
        else:
            colours.append(-1)
    else:
        colours.append(-1)
col_ser = pd.Series(data = colours, index = all_products.index)
```

```python
all_products['major_colour'] = col_ser
all_products.head()
```

```python
price = []
for product in all_products['product_name']:
    instance = df_orders.loc[df_orders['name'] == product]
    price.append(instance['price'].iloc[0])
price_series = pd.Series(data = price, index = all_products.index)
```

```python
all_products['price'] = price_series
all_products.head()
```

```python
#remove products that have a sale of only 1 or less
selling_products = all_products[all_products.no_sold > 1]
selling_products.head()
```

```python
trend = monthly_sales.total - monthly_detrend
```

```python
monthly_trends = []
for i in range(12):
    monthly_trends.append(monthly_detrend[i::12].mean())
```

```python
plt.plot(monthly_trends)
```

```python
deseasoned = monthly_detrend - (monthly_trends * 50)[:len(monthly_detrend)]
```
